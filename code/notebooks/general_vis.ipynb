```python
needed_1 = {
    'random_seed': [101],
    'train_window': [10],
    'key_a': ["default"],
    'key_w': ["default", "simple_101", "const"]
}
```

```python
def extract_needed(experiments, needed):
    vital = []
    for experiment in experiments:
        if experiment.random_seed not in needed['random_seed']:
            continue
        if experiment.train_window not in needed['train_window']:
            continue
        if experiment.key_a not in needed['key_a']:
            continue
        if experiment.key_w not in needed['key_w']:
            continue
        vital.append(experiment)
    return vital
```

```python
vital = extract_needed(experiments_10 + experiments_11, needed_1)
anls = [experiment.logs for experiment in vital]
labels = [experiment.key_a + " + " + experiment.mixing_type + " + " + experiment.key_w for experiment in vital]
colors = [None] * len(vital)
draw_several(from_start=False, logs=anls, labels=labels, colors=colors, 
             best_color="black", best_label="Best partition",
             title=None, fig_size=(14, 8), loc="center left", save_path=None) #"foo.eps")
```
